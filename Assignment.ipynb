```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("C:/Users/user/Desktop/titanic.csv")

print("\n", df[["Age", "Sex"]].head())

plt.figure(figsize=(9,6))

sns.histplot(data=df, x="Age", hue="Sex", bins=50, kde=False)

plt.title("Age Distribution of Titanic Passengers (Male vs Female)")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```
